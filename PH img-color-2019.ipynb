```python
def create_palette_img(palette, img_name, folder): #create the image from the colors array
    print(folder)
    import numpy as np
    import os
    import scipy.misc 
    
    # Image size
    width = 100
    height = 100
    channels = 3
    
    # Create an empty image
    img = np.zeros((height, width, channels), dtype=np.uint8)
    
    print(palette)
    
    # Set the RGB values
    #colora i pixel dell'imagine con i colori del dizionario "palette"
    for y in range(img.shape[0]):
        for x in range(img.shape[1]):
            img[y][x][0] = palette[0]
            img[y][x][1] = palette[1]
            img[y][x][2] = palette[2]

    if not os.path.exists(folder):
        os.makedirs(folder)        
            
    # Save the image
    scipy.misc.imsave(folder+"/{}.png".format(img_name), img)
    
```

```python
def palette_mosaic(palette3, img_name): 
    import numpy as np
    import scipy.misc 

    # Image size
    height = len(palette3)*10
    width = 160
    channels = 3

    # Create an empty image
    img = np.zeros((height, width, channels), dtype=np.uint8)

    for i in range(width):
        for y in range(height):
            row = int(y/40)
            indice = int((row*4)+(i/40)) #i/100 = 0-3 
            r, g, b = palette3[indice]

            img[y][i][0] = r
            img[y][i][1] = g
            img[y][i][2] = b
    

    # Save the image
    scipy.misc.imsave("phImgColor2019/palette/%s.png" %img_name, img)
    print("mosaic %s created" %img_name)
```

```python
def createPalette(month, year): #detect colors in screenshots
    import colorsys
    from colorthief import ColorThief
    import os
    
    directory = "phImgColor2019/ph-imgs/"+year+"-"+month
    palette_dir = "phImgColor2019/palette/"+year+"-"+month
    screenshots_dir = os.listdir(directory) #list all files in the selected directory

    palette3 = []
    palette2 = []
    
    #get color palette
    for screenshot in screenshots_dir:
        if "jpg" in screenshot:           
            img_temp = directory+'/%s' % screenshot
            print(img_temp)
            color_thief = ColorThief(img_temp)
            # get the dominant color
            dominant_color = color_thief.get_color(quality=5) #quality 1 max - 10 min
            # build a color palette
            palette = color_thief.get_palette(color_count=4)
        
            palette3.extend(palette)

            create_palette_img(dominant_color, 'palette-'+str(screenshot[:-4]), palette_dir) #mia funzione per create l'img 
    
    palette3.sort(key=lambda rgb: colorsys.rgb_to_hls(*rgb))
    palette_mosaic(palette3, "palette"+year+"-"+month)
```
